```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/content/gdrive/My Drive/Character Analysis.xlsx")
theme_counts = df["Nietzschean Themes"].value_counts()
most_repeated_themes = theme_counts.nlargest(5)  # Get the top 5 themes
print(f"Most repeated Nietzschean Themes: {most_repeated_themes}")
```

```python
most_repeated_themes.plot(kind="bar")
plt.xlabel("Theme")
plt.ylabel("Frequency")
plt.title("Frequency of Most Repeated Nietzschean Themes")
plt.show()
```

# Bar Graph Analysis

Wierdly enough, **Will To Truth is the most prominent one**, standing out exponentially taller compared to the rest. It is then followed by other themes which oddly enough share an equal distribution.
As Nietzsche once quoted, **"If you wish to strive for peace of soul and happiness, then believe; if you wish to be a disciple of truth, then inquire."** The Will to Truth is a dominant struggle, which is also the struggle for Freedom. As evident from the analysis of the characters, especially the ones following existentialist philosphy, Will To Truth is a driving factor as existentialism mainly deals with finding meaning in the world. **That meaning then is a function of the incessant search for the Truth**.

```python
import seaborn as sns

theme_counts_by_anime = pd.pivot_table(df, values="Nietzschean Themes", index="Anime Theme", aggfunc="count")
sns.heatmap(theme_counts_by_anime, annot=True)
plt.title("Frequency of Nietzschean Themes by Anime Theme")
plt.show()
```

```python
character_connections = df.to_dict(orient='index')
for character, connections in character_connections.items():
    del connections['Character Name']
    for connected_character, theme_count in connections.items():
        connections[connected_character] = theme_count
```

```python
for character, connections in character_connections.items():
    for connected_character, weight in connections.items():
        try:
            float(weight)
        except ValueError:
            print(f"Invalid weight: {character} -> {connected_character} = {weight}")
```

```python
import networkx as nx
G = nx.Graph()
for character, connections in character_connections.items():
    for connected_character, weight in connections.items():
        try:
            weight = float(weight)
        except ValueError:
            weight = 0.0
        G.add_edge(character, connected_character, weight=weight)
```

```python
nx.draw_networkx(G, with_labels=True)
```
